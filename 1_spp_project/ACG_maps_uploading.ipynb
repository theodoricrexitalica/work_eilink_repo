# Data uploading

```python
def maps_uploading(folder_path):
    file_names = os.listdir(folder_path)
    df_list = []
    for items in tqdm(file_names):
        file_path = os.path.join(folder_path, items)
        csv = pd.read_csv(file_path)
        if 'VIII' in file_path:
            csv['FORMATION_up'] = 'Balakhany VIII'
        if 'X' in file_path:
            csv['FORMATION_up'] = 'Balakhany X'
        df_list.append(csv)
    maps = pd.concat(df_list)
    if 'x/y/NET' in maps.columns:
        maps = maps.drop('x/y/NET', axis=1)
    elif 'x/y/PHIT' in maps.columns:
        maps = maps.drop('x/y/PHIT', axis=1)
    else:
        pass

    return maps
```

```python
def cut_azery_ntg_bal8(maps_ntg, dataset, margin_well):
    coef = 500
    Xmin = dataset.Xmean.min() - margin_well
    Xmax = dataset.Xmean.max() + margin_well
    Ymin = dataset.Ymean.min() - margin_well
    Ymax = dataset.Ymean.max() + margin_well

    formations = [x for x in maps_ntg.FORMATION.unique() if x is not None]

    fm_list8 = []
    for fm in formations:
        if 'VIII' in fm:
            maps_ntg_viii = maps_ntg[maps_ntg.FORMATION == fm]
            maps_ntg_viii = maps_ntg_viii[  (maps_ntg_viii.Xmap >= Xmin) &
                                            (maps_ntg_viii.Xmap <= Xmax) &
                                            (maps_ntg_viii.Ymap >= Ymin) &
                                            (maps_ntg_viii.Ymap <= Ymax)]
            fm_list8.append(maps_ntg_viii)
    return fm_list8
ntg_fm_list8 = cut_azery_ntg_bal8(maps_ntg, df_bal8_azr_v2_flag, margin_well=500)

def map_ntg_display_bal8(datalist, well_coord):
    colors = ['#029616', '#e9e6d9', '#e9da0c', ]
    cmap = LinearSegmentedColormap.from_list("", colors)
    for df_fm in datalist:
        fig, ax = plt.subplots(figsize=(12, 7))
        cf = ax.scatter(df_fm['Xmap'], df_fm['Ymap'], c=df_fm['NTGmap'], cmap=cmap, s=10, vmin=0, vmax=1)
        ax.scatter(well_coord['Xmean'], well_coord['Ymean'], c='yellow', s=50, alpha=0.5, ec='black')
        ax.scatter(well_coord['Xmean'], well_coord['Ymean'], c='black', s=5)
        # for ind, row in well_coord.iterrows():
        #     if ind % 2 == 0:
        #         ax.text(row['Xmean'], row['Ymean'], row['well'], fontsize=4, c='black', ha='left', va='bottom')
        #     else:
        #         ax.text(row['Xmean'], row['Ymean'], row['well'], fontsize=4, c='black', ha='left', va='bottom')
        ax.set_title(df_fm.columns[2] + ' ' + str(df_fm['FORMATION'].iloc[0]))
        fig.colorbar(cf, ax=ax, label='ntg')
        fig.show()
        # plt.savefig(df_fm.columns[2] + ' ' + str(df_fm['FORMATION'].iloc[0]) + '.pdf', dpi=600, bbox_inches='tight');
map_ntg_display_bal8(ntg_fm_list8, well_coord8)
```

```python
def histoplot_ntg_phit(datalist, var, comment, bins=50):
    df = pd.concat(datalist)
    categories = df['FORMATION'].unique()
    for category in categories:
        sns.histplot(df[df['FORMATION'] == category][var], kde=True, bins=50, label=category)
    plt.legend(title=comment)
    plt.show()
histoplot_ntg_phit(ntg_fm_list8, 'NTGmap', 'NTG Bal8')
```

# Wells ngt vs maps ngt

```python
def ntg_avg_calculation(dataset_wells):
    well_fm_ngt = dataset_wells.groupby(['well','FORMATION'])[['NET','TST']].agg(
                {'NET': lambda x: x.sum()*0.1,'TST': lambda x: x.iloc[-1] - x.iloc[0]}
                ).reset_index()
    well_fm_ngt = well_fm_ngt.rename(columns={'NET':'NETsum','TST':'TST_interv'})
    well_fm_ngt['NGT_avg'] = well_fm_ngt.NETsum / well_fm_ngt.TST_interv
    return well_fm_ngt

def ntg_avgmap_calculation(dataset_wells, maps_var, zone_size):
    def formation_join(dataset_wells, maps_var):
        well_fm_xy = dataset_wells.groupby(['well','FORMATION'])[['X_traj','Y_traj']].mean().reset_index()
        well_fm_xy = well_fm_xy.rename(columns={'X_traj':'Xmean', 'Y_traj':'Ymean'})
        selected_fm = []
        for fm1 in well_fm_xy.FORMATION.unique():
            if fm1 in maps_var.FORMATION.unique():
                selected_fm.append(fm1)
        return selected_fm, well_fm_xy
    selected_fm, well_fm_xy = formation_join(dataset_wells, maps_var)

    def map_var_wells(well_coords, maps_var, fm, zone_size, var):
        # print(f'Calc ntg_map data for {fm}')
        well_coords_fm = well_coords[well_coords.FORMATION == fm]
        maps_var_fm = maps_var[maps_var.FORMATION == fm]
        df_lst = []
        for wellname in well_coords_fm.well.unique():
            data_fm = well_coords_fm[(well_coords_fm.well == wellname)]
            well_x = data_fm['Xmean'].values[0].round(0)
            well_y = data_fm['Ymean'].values[0].round(0)
            map_avg = maps_var_fm[  (maps_var_fm.Xmap < well_x + zone_size) &
                                    (maps_var_fm.Xmap > well_x - zone_size) &
                                    (maps_var_fm.Ymap < well_y + zone_size) &
                                    (maps_var_fm.Ymap > well_y - zone_size)].reset_index(drop=True)
            df_avg = map_avg[var].mean()
            df = pd.DataFrame({'well': [wellname], 'FORMATION':[fm], 'Xmean': [well_x], 'Ymean': [well_y], 
                            '_'.join([var, str(zone_size)]): [df_avg]})
            df_lst.append(df)
        maps_ntg_wells_fm = pd.concat(df_lst).reset_index(drop=True)
        maps_ntg_wells_fm = maps_ntg_wells_fm.rename(columns={var:var + '_avg'})
        return maps_ntg_wells_fm
    df_fm_list = []
    for fm in selected_fm:
        maps_wells_fm = map_var_wells(well_fm_xy, maps_var, fm, zone_size, 'NTGmap')
        df_fm_list.append(maps_wells_fm)
    maps_ntg_wells = pd.concat(df_fm_list).reset_index(drop=True)
    return maps_ntg_wells

def wells_map_joining(dataset_wells_res, maps_var_res):    
    map_wells_ngt_join = maps_var_res.set_index(['well','FORMATION']).join(dataset_wells_res.set_index(['well','FORMATION'])).reset_index()
    map_wells_ngt_join = map_wells_ngt_join[(map_wells_ngt_join.NGT_avg < 1) & 
                                            (map_wells_ngt_join.NGT_avg > 0)]
    return map_wells_ngt_join

def ntg_xplots_display(dataset_wells_map, zone_size):
    hex_colors = [mcolors.to_hex(c) for c in sns.color_palette()]
    fm_list = [fm for fm in dataset_wells_map.FORMATION.unique()]
    target_map_names = '_'.join(['NTGmap', str(zone_size)])

    fig, ax = plt.subplots(1,len(fm_list), figsize=(len(fm_list)*4.5,5))
    for idx, fm in enumerate(fm_list):
        data = dataset_wells_map[dataset_wells_map.FORMATION ==  fm]

        sns.scatterplot(data=data, 
                        x=target_map_names, y='NGT_avg', color=hex_colors[idx], ec='black', alpha = 0.66, ax=ax[idx])
        sns.regplot(data=data, x=target_map_names, y='NGT_avg', scatter=False, color='red', ax=ax[idx])
        sns.lineplot(x=[0,1], y=[0,1], color='black', ax=ax[idx], ls='--')
        ax[idx].text(0.1,0.98, f'corr_coef: {data[target_map_names].corr(data["NGT_avg"]).round(2)}', fontsize=11)
        ax[idx].set_title(fm_list[idx] + ' size:' + str(zone_size) + 'm')
    fig.show()

zone_size_list = [25, 50, 100]
result_list_ntg = []
for zone_size in zone_size_list:
    print('Calculation ntg-map data with zone size: ', zone_size, 'm')

    well_fm_ngt8 = ntg_avg_calculation(df_bal8_azr_v2_flag)
    well_fm_ngt10 = ntg_avg_calculation(df_bal10_vshclp2_flag)

    maps_ntg_wells8 = ntg_avgmap_calculation(df_bal8_azr_v2_flag, maps_ntg, zone_size)
    maps_ntg_wells10 = ntg_avgmap_calculation(df_bal10_vshclp2_flag, maps_ntg, zone_size)

    map_wells_ngt_join8 = wells_map_joining(well_fm_ngt8, maps_ntg_wells8)
    map_wells_ngt_join8['FORMATION_up'] = 'Balakhany VIII'
    map_wells_ngt_join10 = wells_map_joining(well_fm_ngt10, maps_ntg_wells10)
    map_wells_ngt_join10['FORMATION_up'] = 'Balakhany X'

    final_ngt_map_wells = pd.concat([map_wells_ngt_join8, map_wells_ngt_join10]).sort_values('well')
    result_list_ntg.append(final_ngt_map_wells)

def merge_data(data_list, zone_size_list):
    merged_df = data_list[0]
    for df in data_list[1:]:
        merged_df = merged_df.merge(df.drop([   'Xmean', 'Ymean', 'NETsum',
                                                'TST_interv', 'NGT_avg', 'FORMATION_up'], axis=1), on=['well', 'FORMATION'], how='outer')
    cols = merged_df.columns.tolist()
    cols.insert(5, cols.pop(cols.index('NTGmap_' + str(zone_size_list[1])))) 
    cols.insert(6, cols.pop(cols.index('NTGmap_' + str(zone_size_list[2]))))
    merged_df = merged_df.reindex(columns=cols)
    return merged_df

merged_df_ntg = merge_data(result_list_ntg, zone_size_list)
merged_df_ntg.to_csv(r'C:\jupyter\spp_farid_data\phit_ntg_maps\final_ngt_map_wells.csv', index=False)
```
